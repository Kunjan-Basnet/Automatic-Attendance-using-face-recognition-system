```python
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
import time
import cv2
from datetime import datetime
```

```python
data={
    "name":["Aaditya"],
    "attendence":[0],
    "attendence_date":["00"]
}
```

```python
df=pd.DataFrame(data)
```

```python
df.to_csv("attendence.csv")
```

```python
classifier=load_model('models/classifierfive.h5')
```

```python
bounding=load_model('models/boundingfive.h5')
```

```python
classification_dict={
    0:"Aaditya",
    1:"No one"
    
}
```

```python

counter=0
cap = cv2.VideoCapture(0)
while cap.isOpened():
    _ , frame = cap.read()
    #frame = frame[100:800, 400:1200,:]
    frame=cv2.resize(frame,(224,224))
    
    
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    resized = tf.image.resize(rgb, (120,120))
    
    yhat = classifier.predict(np.expand_dims(resized/255,0))
    class_text=classification_dict[np.argmax(yhat)]
    if class_text!="No one":
        now = datetime.now()
        current_hour = now.minute

        datafr=pd.read_csv("attendence.csv")
        x=datafr.loc[datafr['name']==class_text,'attendence_date'].values
        y=int(x[0])
        if y<int(current_hour):
            datafr.loc[df['name'] == class_text, 'attendence'] += 1
            nowo = datetime.now()
            current_houro = nowo.minute


            datafr.loc[df['name'] == class_text, 'attendence_date'] = str(current_houro)
            datafr.to_csv("attendence.csv",index=False)
    
    
    
    coords=bounding.predict(np.expand_dims(resized/255,0))
    x1=int(coords[0][0]*224)
    y1=int(coords[0][1]*224)
    x2=int(coords[0][2]*224)
    y2=int(coords[0][3]*224)
    
    cv2.rectangle(frame, (x1,y1), (x2,y2), (255,0,0), 3)
    
    cv2.putText(frame, class_text, (x1, y1 + 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    cv2.imshow('Video', frame)
    
    
    

    
    


    
    
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
