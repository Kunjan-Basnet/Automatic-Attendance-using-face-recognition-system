```python
import albumentations as alb

augmentor = alb.Compose([
    alb.Resize(height=224, width=224, p=1),
    alb.RandomBrightnessContrast(p=0.2),
    alb.RandomGamma(p=0.2),
    alb.RGBShift(p=0.2),
    alb.Rotate(limit=90, p=0.5)
],
bbox_params=alb.BboxParams(format='albumentations', label_fields=['class_label']))
#yaa class_label lekhya vayae ra tala class_label vanni nai use gare ra yesma pathauni
#matab yo class_label ma chai ma class ko label pathauxu vane ko
```

```python
for folders in ['train','valid']:
    for images in os.listdir(os.path.join('data',folders,'images')):
        if images.split('.')[1]=='jpg':
            image_name=images
            label_name=image_name.split('.')[0]+'.json'
            print(image_name)
            print(label_name)
            #extract coordinates and normalize the labels
            if os.path.exists(os.path.join('data',folders,'labels',label_name)):
                with open(f'data/{folders}/labels/{label_name}','r') as f:
                    content=json.load(f)
                    #print(content.keys())
                    #print(content['shapes'][0]['points'])
                    coordinates=content['shapes'][0]['points']
                    #coordinates in matrix form
                    x1=coordinates[0][0]
                    y1=coordinates[0][1]
                    x2=coordinates[1][0]
                    y2=coordinates[1][1]
                    normalized_coords=[x1/content['imageWidth'],y1/content['imageHeight'],x2/content['imageWidth'],y2/content['imageHeight']]
                    #print(normalized_coords)
                    #print(content['imageWidth'])


                    #get label
                    my_class=content['shapes'][0]['label']

            img=cv2.imread(os.path.join('data',folders,'images',image_name))
            for x in range(30):
                print(image_name)
                augmented = augmentor(image=img, bboxes=[normalized_coords], class_label=[my_class])
                #print(type(augmented))
                #print(augmented.keys())
                #print(augmented['bboxes'])
                cv2.imwrite(os.path.join('aug_data', folders, 'images', f'{image_name.split(".")[0]}.{x}.JPG'), augmented['image'])




                #aaba label haru baauni yesko lagi
                #json format ma banauni
                if os.path.exists(os.path.join('data',folders,'labels',label_name)):
                    if(len(augmented['bboxes'])==0):
                        #yedi tauko image bhaira paryo vanae augmented['bboxes']=[] aauxa vanae ko khali array aauxa
                        annotation={}
                        print(augmented['bboxes'])
                        annotation['bbox']=[0,0,0,0]
                        annotation['imageWidth']=224
                        annotation['imageHeight']=224
                        annotation['class']=['noone']
                        print(annotation)

                    else:


                        annotation={}
                        print(augmented['bboxes'])
                        annotation['bbox']=augmented['bboxes'][0]
                        annotation['imageWidth']=224
                        annotation['imageHeight']=224
                        annotation['class']=augmented['class_label'][0]
                        print(annotation)
                else:
                        annotation={}
                        print(augmented['bboxes'])
                        annotation['bbox']=[0,0,0,0]
                        annotation['imageWidth']=224
                        annotation['imageHeight']=224
                        annotation['class']=['noone']
                        print(annotation)
                                  


                # Save the annotation JSON file
                annotation_path = os.path.join('aug_data', folders, 'labels', f'{image_name.split(".")[0]}.{x}.json')
                with open(annotation_path, 'w') as f:
                    json.dump(annotation, f)
                    pass


            
                        
            
            
            
            
            
            
            
```
