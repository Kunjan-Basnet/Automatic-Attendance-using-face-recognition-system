```python
import os
import random
```

```python
for folders in os.listdir('data'):
    if os.path.isdir(os.path.join('data',folders)):
        if folders=='images':
            total=len(os.listdir('data/images'))
            print(total)
            count=0
            images_folder=os.listdir(f'data/{folders}')
            random.shuffle(images_folder)
            for images in images_folder:
                
                image_name=images
                label_name=image_name.split('.')[0]+'.json'
                print(image_name)
                print(label_name)
                if(count<int(total*0.8)):
                    print("for training")
                    print(count)
                    os.rename(os.path.join('data',folders,image_name),os.path.join('data','train','images',image_name))
                    if os.path.exists(os.path.join('data','labels',label_name)):
                        os.rename(os.path.join('data','labels',label_name),os.path.join('data','train','labels',label_name))
                    
                    count +=1
                    
                else:
                    print("for validation")
                    print(count)
                    os.rename(os.path.join('data',folders,image_name),os.path.join('data','valid','images',image_name))
                    if os.path.exists(os.path.join('data','labels',label_name)):
                        os.rename(os.path.join('data','labels',label_name),os.path.join('data','valid','labels',label_name))
                    count +=1
                    
                
                
            
```
